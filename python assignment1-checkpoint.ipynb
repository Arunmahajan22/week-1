```python
import random as r
l=[r.randint(100,900) for i in range(100)]
print(l)
countodd=0
counteven=0
for i in l:
    if i%2!=0:
        countodd+=1
    else:
        counteven+=1
print("total evennumbers",counteven)
print("total oddnumbers",countodd)
countprime=0
flag=0
for i in l:
    for j in range(2,i//2+1):
        if i%j==0:
            flag=1
    if flag==0:
        countprime+=1
print("total primeno.",countprime)
```
